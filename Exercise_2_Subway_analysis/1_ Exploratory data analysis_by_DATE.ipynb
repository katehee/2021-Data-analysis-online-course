> ###  Q) Are there higher number of riders found in May?

> ### Q) Is Friday the busiest day of the week? 

### Hypothesis1)   Are there higher number of riders found in May?

```python
#create 'month' col
yearmonth_list = []
monthday_list = []
for date in raw['Date']:
    yearmonth = str(date)[:6]   # 왼쪽부터 6자리 문자 선택
    yearmonth_list.append(yearmonth)
    monthday = str(date)[4:]    # 왼쪽에서 5번째 문자부터 끝까지 선택
    monthday_list.append(monthday)
    
# 연월/ 월일 컬럼 추가하기
raw['yearmonth'] = yearmonth_list
raw['monthday'] = monthday_list
raw.head()
```

```python
#user pivot to find the no. rider in each month
data_month = pd.pivot_table(raw, index = 'yearmonth', values = 'On-board', aggfunc='sum')
data_month = data_month.sort_values(by = 'On-board', ascending= False)
data_month.head()
```

### Hypothesis1)   Are there higher number of riders found in May? ---> `True`
