# Mapping
- Create a map
    - `m` = folium.Map(location = [latitude, longitude],  zoom_start = zoom scale)

- Add information
    - Add a marker
        - folium.Marker([latitude, longitude]).add_to(`m`)
    - Add a circle
        -     folium.CircleMarker([latitude, longitude],radius= circle size).add_to(`m`)
    - Additional options: 
        - tooltip="cursor message"
        - popup="message to be appeared when clicked"          
    - ex) ClickForMarker('CheckPoint').add_to(`m`) 

```python
# Create a map
# Set latitude and longitude boundary  
# zoom_start: set zoom scale 
m = folium.Map(location=[43.653963,-79.387207],   #Toronto)
               zoom_start=12)
m
```

```python
#Add a marker
folium.Marker([43.653963,-79.387207],    # Toronto location
              tooltip="Toronto(cursor msg)",
              popup="Toronto(clicked msg)",
              ).add_to(m)
m
```

```python
# Circle marker 
folium.CircleMarker([43.653963,-79.387207],
                    radius=20,
                    tooltip = 'cursor msg'
               ).add_to(m)
m
```

## Minimap

```python
from folium.plugins import MiniMap

# Create a map
m = folium.Map(location=[43.653963,-79.387207],   # 기준좌표: 서울역
               zoom_start=12)

# Create a Mini map
minimap = MiniMap() 
minimap.add_to(m)


# Add a marker
folium.Marker([43.653963,-79.387207],    # 서울역위치
#               tooltip="Toronto(cursor)",
#               popup="Toronto(clicked)",
              ).add_to(m)
m
```

### folium.ClickForMarker : Add markers by clicking on the map

```python
# Create a map
m = folium.Map(location=[43.653963,-79.387207],   # 기준좌표: 서울역
               zoom_start=12)

# Add a circle marker
folium.CircleMarker([43.653963,-79.387207],
                    radius=20,
#                     tooltip = 
               ).add_to(m)
folium.ClickForMarker('Marker').add_to(m)    
m
```

```python
# Create a map, then add markers
m = folium.Map(location=[33.68357027447924, 17.78557297949212],
               zoom_start=1)
m
# 대피소 마커 추가하기

for i in range(len(raw)):
    lat = raw.loc[i,'Latitude']
    long = raw.loc[i,'Longitude']
    country = raw.loc[i,'Country']

    folium.Marker([lat, long],tooltip= country).add_to(m)    
m
```
