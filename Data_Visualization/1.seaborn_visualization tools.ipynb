## Data distribution (Numeric Vs Numeric)

### `relplot`( data = df, x = ,  y = , hue = , kind = 'scatter')
relationship between two numeric attributes
- kind option
    - 'scatter' default
    - 'line'

```python
sns.relplot(x = 'tip', y = 'total_bill', data = raw) 
```

```python
#kind option to 'line'
sns.relplot(x = 'tip', y = 'total_bill', data = raw, kind = 'line') 
```

```python
#gender difference 
#hue option to differentiate gender
sns.relplot(x = 'tip', y = 'total_bill', data = raw, hue = 'sex')
```

## Data Relationship (Numeric Vs Numeric)

### `jointplot`( data = df, x = ,  y = , kind = 'scatter')


- kind option
    - "scatter"(default): point   
    - "reg": point + regression
    - "kde": probability density function of a random variable

```python
# jointplot allows to observe relationship and distribution of two attributes
sns.jointplot(data = raw, x = 'tip', y = 'total_bill')  # kind      kind = 'scatter'
```

```python
# kind = 'kde': visualization of density function
# darker = densed 
# fill=True : color fill
sns.jointplot(data = raw, x = 'tip', y = 'total_bill', kind = 'kde', fill=True)  
```

```python
# kind = 'reg' : regression line
sns.jointplot(data = raw, x = 'tip', y = 'total_bill', kind = 'reg')  
```

```python
# kind = 'hex' : similar to kde but areas divided with hexagons
sns.jointplot(data = raw, x = 'tip', y = 'total_bill', kind = 'hex')  
```

### `pairplot`( data = df)

```python
#total_bill, tip, size data relationship through pairplot
sns.pairplot(data = raw)
```

```python
#gender difference
sns.pairplot(data = raw, hue = 'sex')
```

## Data distribution(Numeric Vs Categoric Vs Categoric)
### `heatmap`( data = df )

```python
df = raw.pivot_table(index = 'day', columns = 'size', values = 'tip', aggfunc='mean')
df
```

```python
sns.heatmap(data = df)
```

```python
sns.heatmap(data = df, 
           annot = True, fmt = '.2f')
```

```python
sns.heatmap(data = df, 
           annot = True, fmt = '0.2f',
           cmap = 'Blues')
```
